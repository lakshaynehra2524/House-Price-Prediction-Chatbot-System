```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def normalize_text(text):
    tokens = word_tokenize(text.lower())
    clean_tokens = []
    for t in tokens:
        if t.isalnum() and t not in stop_words:
            clean_tokens.append(lemmatizer.lemmatize(t))
    return clean_tokens
```
